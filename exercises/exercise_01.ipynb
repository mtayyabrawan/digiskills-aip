## Task 1 (Marks 03)

Write a program that takes salary as input. Using conditional statements, calculate the final tax
based on the following rules:

- If the salary is less than 30,000 → Tax rate is 5%
- If the salary is between 30,000 and 70,000 → Tax rate is 15%
- If the salary is greater than 70,000 → Tax rate is 25%

```python
def calculate_tax(salary, tax_percentage):
    return (salary * tax_percentage) / 100


try:
    employee_salary = float(input("Enter your salary (only numbers) :"))
    final_tax = None
    if employee_salary <= 0:
        final_tax = calculate_tax(employee_salary, 0)
    elif employee_salary < 30000:
        final_tax = calculate_tax(employee_salary, 5)
    elif employee_salary >= 30000 and employee_salary <= 70000:
        final_tax = calculate_tax(employee_salary, 15)
    else:
        final_tax = calculate_tax(employee_salary, 25)
    print(
        f"Your salary after applying tax of Rs. {final_tax} is Rs. {employee_salary-final_tax}"
    )
except ValueError:
    print("Salary can only be numerical value")
```

## Task 2 (Marks 03)

Given a list of words:

`words = ["apple", "banana", "kiwi", "cherry", "mango"]`

Create a dictionary that maps each word to its corresponding length.

Example Output: `{"apple": 5, "banana": 6, "kiwi": 4, "cherry": 6, "mango": 5}`

```python
words = ["apple", "banana", "kiwi", "cherry", "mango"]

words_dict = dict()

for word in words:
    words_dict[word] = len(word)

print(words_dict)
```

## Task 3 (Marks 04)

Create a Chat System using Object-Oriented Programming (OOP) concepts. (Marks 4)
You need to create the following classes:

- User
- Message
- ChatRoom 

The system should implement the following functionalities:

- Sending messages
- Viewing chat history
- User joining and leaving the chat room

```python
from datetime import datetime
import random
from typing import List, Union


def random_id():
    return random.randrange(10000000, 99999999)


class User:
    def __init__(self, username):
        self.username = username
        self.created_at = datetime.now()
        self.user_id = random_id()

    def __str__(self):
        return f"User Id: {self.user_id}\nUsername: {self.username}\nCreated At: {self.created_at}"


class Message:
    def __init__(self, msg, user_id):
        self.message_id = random_id()
        self.message = msg
        self.user_id = user_id
        self.timestamp = datetime.now()

    def __str__(self):
        return f"{self.user_id}: {self.message} ({self.timestamp})"


class ChatRoom:
    __users = set()
    messages = []

    def __init__(self, name):
        self.name = name
        self.chatroom_id = random_id()

    def join(self, user_id):
        self.__users.add(user_id)

    def leave(self, user_id):
        self.__users.discard(user_id)

    def send_message(self, message: Message):
        if self.__users.issuperset({message.user_id}):
            self.messages.append(message)
        else:
            print("You can't send message as you are not the chatroom member")

    def chat_history(self):
        for message in self.messages:
            print(message)

    def __str__(self):
        return f"Chatroom Id: {self.chatroom_id}\nChatroom Name: {self.name}"


def menu():
    print("Welcome to Chatpy")
    user_input = None
    try:
        print(
            "(1) Create a User\n(2) Create a chatroom\n(3) List Chatrooms\n(4) List Users\n(5) Switch user\n(6) Join chatroom\n(7) Leave chatroom\n(8) Send message\n(9) Chat history\n(10) Exit"
        )
        user_input = int(float(input("Choose from the menu: ")))
    except ValueError as _:
        print("Only integer values ar accepted")
    return (
        user_input
        if (type(user_input) == int and (user_input > 0 and user_input < 11))
        else menu()
    )


users: List[User] = []
chatrooms: List[ChatRoom] = []
current_user = None
current_chatroom = None


def create_user():
    username = input("Enter New username: ")
    new_user = User(username)
    global current_user
    current_user = new_user
    users.append(new_user)
    print(f"Welcome to Chatpy! [{new_user.user_id}]:{new_user.username}")


def create_chatroom():
    chatroom_name = input("Enter Chatroom name: ")
    new_chatroom = ChatRoom(chatroom_name)
    global current_chatroom, current_user
    if current_chatroom == None and current_user != None:
        current_chatroom = new_chatroom
        current_chatroom.join(current_user.user_id)
        print("You successfully joined this chatroom")
    chatrooms.append(new_chatroom)
    print(f"Chatroom created! successfully")


def list_chatrooms():
    for chatroom in chatrooms:
        print(chatroom)


def list_users():
    for user in users:
        print(user)


def switch_user():
    user_id = None

    def check_user(user: User):
        return user.user_id == user_id

    try:
        user_id = int(input("Enter your user id to switch :"))
    except ValueError as _:
        print("Only integer values ar accepted")
    if user_id == None:
        return switch_user()
    else:
        filtered_users = list(filter(check_user, users))
        if len(filtered_users) == 0:
            print(f"User with User Id :{user_id} not found")
            return
        else:
            print("User switched successfully")
            global current_user
            current_user = filtered_users[0]


def join_chatroom():
    chatroom_id = None

    def check_chatroom(chatroom: ChatRoom):
        return chatroom.chatroom_id == chatroom_id

    try:
        chatroom_id = int(input("Enter chatroom id to join :"))
    except ValueError as _:
        print("Only integer values ar accepted")
    if chatroom_id == None:
        return join_chatroom()
    else:
        filtered_chatrooms = list(filter(check_chatroom, chatrooms))
        if len(filtered_chatrooms) == 0:
            print(f"Chatroom with Chatroom Id :{chatroom_id} not found")
            return
        else:
            global current_chatroom, current_user
            current_chatroom = filtered_chatrooms[0]
            if current_user == None:
                create_user()
            current_chatroom.join(current_user.user_id)  # type: ignore
            print("Chatroom joined successfully")


def leave_chatroom():
    global current_chatroom, current_user
    if current_user == None:
        print("You don't have any account")
        return
    if current_chatroom == None:
        print("You are not member of any chatroom")
    else:
        current_chatroom.leave(current_user.user_id)
        current_chatroom = None


def send_message():
    global current_user, current_chatroom
    if current_user == None:
        print("You don't have any account. Go first create user account")
    elif current_chatroom == None:
        print("You are not member of any chatroom")
    else:
        message_content = input("Enter your message: ")
        message = Message(message_content, current_user.user_id)
        current_chatroom.send_message(message)
        print("Message sent successfully")


def chat_history():
    global current_user, current_chatroom
    if current_user == None:
        print("You don't have any account. Go first create user account")
    elif current_chatroom == None:
        print("You are not member of any chatroom")
    else:
        print("Chat History")
        current_chatroom.chat_history()


def main():
    print(current_chatroom)
    print(current_user)
    menu_option = menu()
    if menu_option == 10:
        return
    elif menu_option == 1:
        create_user()
    elif menu_option == 2:
        create_chatroom()
    elif menu_option == 3:
        list_chatrooms()
    elif menu_option == 4:
        list_users()
    elif menu_option == 5:
        switch_user()
    elif menu_option == 6:
        join_chatroom()
    elif menu_option == 7:
        leave_chatroom()
    elif menu_option == 8:
        send_message()
    elif menu_option == 9:
        chat_history()
    return main()


main()
```
