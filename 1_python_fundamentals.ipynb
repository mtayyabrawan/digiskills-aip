## Abstraction

Abstraction means hidding attributes and methods of the class that are not essential to be public.

In python we can hide attributes and methods by prefixing `__` to them.

For example:

```python
    class Student:
        def __init__(self,name,roll_no,password):

```

````python
class Student:
    def __init__(self, name, roll_no, password):
        self.name = name
        self.roll_no = roll_no
        self.__password = password


st1 = Student("Muhammad Tayyab", 1, "12345678")

print(st1.name)
print(st1.roll_no)
print(st1.__password)  # printing __password throws exception as it is private attribute
````

## Inheritance

Inheritance means getting another class attributes. In real world child inherits jeans from their parents. In OOP a class can also inherit attributes and methods from another class as child.

```python
    class Animal:
        def __init__(self,name):
            self.name = name
        def sound(self):
            print(f"{self.name} is making sound")

    class Dog(Animal):
        def __init__(self,name):
            super().__init__(name)
        @staticmethod
        def sound():
            print("Woff Woff!")

    anim = Animal("Animal")
    anim.sound()

    puppy = Dog("Puppy")
    puppy.sound()
```

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} is making sound")


class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

    @staticmethod
    def bark():
        print("Woff Woff!")


anim = Animal("Animal")
anim.sound()

puppy = Dog("Puppy")
puppy.sound()  # .sound() is inherited from Animal class
puppy.bark()
````

## Polymorphism

Polymorphism means overriding the attributes and methods of classes. In this process the previously declared attributes & methods of class are overridden by its child.

For example:

```python
    class Cat(Animal):
        def __init__(self,name):
            super().__init__(name)
        @staticmethod
        def sound(): # this method overrides the .sound() of its parent class (Animal)
            print("Meow Meow!")

    tom = Cat("Tom");

    tom.sound()
```


````python
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name)

    @staticmethod
    def sound():  # this method overrides the .sound() of its parent class (Animal)
        print("Meow Meow!")


tom = Cat("Tom")

tom.sound()
````
