## Input & Output functions in python

### Ouput

To display the output we use print function with the text or any vriable inside the parenthesis after print word

For example:

```python
print("Hello World")
```

In above example we want to display Hello World as output so we have written it with quotes inside the parenthesis

````python
print("Hello World")
````

### Input

To input data from the user we use input function, we pass the argument to the function to display while asking for input and when user inputs the data the function returns that data as string and so we can use that data in our program

For example:

```python
user_name = input("Please enter your username: ")
print("Welcome,",user_name)
```

In above example we have asked user to input his/her username and when user provides input we are stroing that data in a variable named user_name and then welcoming that user with print statement

````python
user_name = input("Please enter your name: ")
print("Welcome,", user_name)
````

## Variables

Variables are the containers in which we can store our data.

For example:

We have a box in whih we are storing toys.
Here toys are the data. And we call that box `Toy box` so that is the variable name.

These variables can store different types of data.

Data Type of variables in python are:

- string
- integer
- float
- boolean
- None

There are also some rules for naming varibales. Here are followings:

1. Can only contain numbers, alphabets and underscore symbol ( _ ).
2. Names should must start with alphabets or underscore ( _ ).
3. Spaces are not allowed in varibale names.
4. As python is case senstive it means that varibales with uppercase letters are different from lowercase ones.

We can create variables in python prividing name of the variable and = sign after that which is caled as assignment operator and after that can provide the data that we want to store in the varibale

For example:

```python
my_name = "Muhamad Tayyab"
```

In above example `my_name` is the varibale name in which we want to store the full name and = sign is the assignment operator to assign data to the variabels and `Muhammad Tayyab` is the data we are assigning to the variable.

````python
my_name = "Muhammad Tayyab Raza Awan"
````

### Checking type of variable

We can check type of varibales by using type() function by passing varibale as argument which returns the data type of the provided variable

For example:

```python
print(type(my_name))
```

In above example we are checking the type of varibale named my_name and so in this case it will displays string as we have stored string data in that variable.

````python
print(type(my_name))
````
