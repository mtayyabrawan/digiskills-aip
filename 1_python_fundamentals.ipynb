## Variables

Variables are the containers in which we can store our data.

For example:

We have a box in whih we are storing toys.
Here toys are the data. And we call that box `Toy box` so that is the variable name.

These variables can store different types of data.

Data Type of variables in python are:

- string
- integer
- float
- boolean
- None

There are also some rules for naming varibales. Here are followings:

1. Can only contain numbers, alphabets and underscore symbol ( _ ).
2. Names should must start with alphabets or underscore ( _ ).
3. Spaces are not allowed in varibale names.
4. As python is case senstive it means that varibales with uppercase letters are different from lowercase ones.

We can create variables in python prividing name of the variable and = sign after that which is caled as assignment operator and after that can provide the data that we want to store in the varibale

For example:

```python

```

In above example `my_name` is the varibale name in which we want to store the full name and = sign is the assignment operator to assign data to the variabels and `Muhammad Tayyab` is the data we are assigning to the variable.

````python
my_name = "Muhammad Tayyab Raza Awan"
````

#### Methods

These are the functions bound to only strings that help you to manipulate strings.

##### Count

This method returns the count of occurances of a sub string in the provided string we can pass the sub string as argument to the method

For example:

```python
    para = "My name is Muhammad Tayyab"
    print(para.count("a"))
```

In above example we are counting the letter `a` in the given para

````python
para = "My name is Muhammad Tayyab"
print(para.count("a"))
````

We can also pass the start and end index of the string to count the sub string from.

For example:

```python
    print(para.count("a", 1, 14))
```

In above example we are trying to count sub string `a` from index 1 to 14 (end index value is not counted)

````python
print(para.count("a", 1, 15))
````

##### Find

This method is used to find a sub string from a specific string, like .count method we also optionaly pass start & end argument to it for finding sub string for specific part of string. On success it will return the first index where that sub string found while on failure it returns -1.

For example:

```python
    print(para.find("Tayyab"))
```

In above example we are trying to find sub string `Tayyab` from the `para` string.

````python
print(para.find("Tayyab"))
````

### 1. List

List data structure is used to store different type of data as a list (in other languages like javascript or cpp lists are known as arrays).

For example:

```python
    student_marks = [48,78,65,39]
    print("Student Marks", student_marks, type(student_marks))
    student_info = ["Muhammad Tayyab",21,2103508,"Computer Science"]
    print("Student Info", student_info, type(student_info))
```

In above example first variable is used to store a student marks in a list.
While in second variable we stored student info like name, age, roll no & major subject

````python
student_marks = [48, 78, 65, 39]
print("Student Marks", student_marks, type(student_marks))
student_info = ["Muhammad Tayyab", 21, 2103508, "Computer Science"]
print("Student Info", student_info, type(student_info))
````

#### Characteristics of Lists

- Lists are iterables.
- They are mutable means that we can change them directly using their index.
- Also supports duplicate values

```python
    print("List before mutation", student_info)
    student_info[0] = "Muhammad Tayyab Raza Awan"
    print("List after mutation", student_info)
```

````python
print("List before mutation", student_info)
student_info[0] = "Muhammad Tayyab Raza Awan"
print("List after mutation", student_info)
````

#### Methods

##### Pop

This method is used to remove last item of list and also returns the last item.

For example:

```python
    print("List before running pop", student_info)
    print(student_info.pop())
    print("List after running pop", student_info)
```

In above example the pop method removes last item of list which in this case is `Computer Science` and prints it.

````python
print("List before running pop", student_info)
print(student_info.pop())
print("List after running pop", student_info)
````

##### Append

This method is used to add new item at last of list.

For example:

```python
    print("List before appending", student_info)
    student_info.append("Attock")
    print("List after appending", student_info)
```

In above example the poappend method adds new item at the end of list which in this case is `Attock`.

````python
print("List before appending", student_info)
student_info.append("Attock")
print("List after appending", student_info)
````
